```python
import sys
sys.path.append('..')
from data.preprocessing import load_and_split


import pandas as pd
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
```

```python
# keeping this for now
def svc_model(feature_set):
    X_train, X_test, Y_train, Y_test = load_and_split(feature_set=feature_set)
    svm = LinearSVC()
    svm.fit(X_train, Y_train)
    Y_pred = svm.predict(X_test)
    print(classification_report(Y_test, Y_pred))
```

```python
svc_model('combined')
```

```python
svc_model('metadata_only')
```

```python
svc_model('text_only')
```
