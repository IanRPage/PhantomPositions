```python
import sys

sys.path.append("..")
from data.preprocessing import load_and_split
from visualize import visualize


from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
```

## Combined (metadata + text data)

```python
X_train, X_test, Y_train, Y_test = load_and_split('combined')
```

```python
svm = LinearSVC(C=0.01, random_state=5)
svm.fit(X_train, Y_train)
Y_pred = svm.predict(X_test)
print(classification_report(Y_test, Y_pred))
```

```python
visualize(svm, 'combined', X_test, Y_test, Y_pred)
```
