```python
import pandas as pd
import sys
sys.path.append('..')
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from data.preprocessing import load_and_split
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint
from visualize import visualize

#Build the RF Model
#References: https://scikit-learn.org/stable/index.html
```

## Combined (metadata + text data)

```python
x_train, x_test, y_train, y_test = load_and_split('combined')
```

```python
clf = RandomForestClassifier(n_estimators = 200, random_state = 5) #creates trees
clf.fit(x_train, y_train) #train model
y_pred = clf.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
visualize(clf, 'combined', x_test, y_test, y_pred)
```
