```python
import pandas as pd
import sys
sys.path.append('..')
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from data.preprocessing import load_and_split

#Build the RF Classifier
#References: https://scikit-learn.org/stable/index.html

x_train, y_train, x_test, y_test = load_and_split(feature_set='combined')
clf = RandomForestClassifier(n_estimators = 100, random_state = 5) #creates trees
clf.fit(x_train, y_train) #train data
y_pred = clf.predict(x_test)

f1 = f1_score(y_test, y_pred)
print(f1)
```
