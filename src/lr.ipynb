```python
import sys

sys.path.append("..")
from data.preprocessing import load_and_split
from visualize import visualize
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
X_train, X_test, Y_train, Y_test = load_and_split('combined')
```

```python
reg = LogisticRegression(C=0.01, class_weight={0: 1, 1: 6}, random_state=5)
reg.fit(X_train, Y_train)
Y_pred = reg.predict(X_test)
print(classification_report(Y_test, Y_pred))
```

```python
visualize(reg, 'combined', X_test, Y_test, Y_pred)
```
